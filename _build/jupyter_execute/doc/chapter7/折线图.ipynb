## ÂØºÂÖ•‰∏éÈ¢ÑËÆæ

ËôΩÁÑ∂Âú® `pandas` ‰∏≠ÂèØ‰ª•Áõ¥Êé•Ë∞ÉÁî® `matplotliab` ËøõË°åÂèØËßÜÂåñÔºå‰ΩÜÊòØ‰æùÊóßÈúÄË¶ÅËøõË°åÁõ∏ÂÖ≥ËÆæÁΩÆÔºå‰æãÂ¶ÇÂ≠ó‰Ωì„ÄÅÁ≤æÂ∫¶Á≠â„ÄÇ

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.rcParams['axes.unicode_minus']=False
````

````python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2000", periods=1000))
ts = ts.cumsum()
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list("ABCD"))
df = df.cumsum()
````

## ÊåáÂÆöÂàóÁªòÂà∂

ÁªòÂà∂ df Á¨¨‰∏ÄÂàóÁöÑÊäòÁ∫øÂõæ

````python
df['A'].plot()
plt.show()
````

Â∞Ü df ÁöÑÂõõÂàóÂàÜÂà´ÊîæÂú®Âõõ‰∏™Â≠êÂõæ‰∏ä

````python
df.plot(subplots=True)
plt.show()
````

## ÂÖ®ÈÉ®Âàó

ÁªòÂà∂ df ÂÖ®ÈÉ®ÂàóÁöÑÊäòÁ∫øÂõæ

````python
df.plot()
plt.show()
````

## Ë∞ÉÊï¥Â§ßÂ∞è

ÁªòÂà∂ df ÂÖ®ÈÉ®ÂàóÁöÑÊäòÁ∫øÂõæÔºåÂ∞ÜÂ§ßÂ∞èË∞ÉÊï¥‰∏∫ 1000*600 Êù•ËÆ©ÂõæÁâáÊõ¥ÁæéËßÇ

````python
df.plot(figsize=(10,6))
plt.show()
````

## ‰øÆÊîπÊ†áÈ¢ò

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÊ∑ªÂä†Ê†áÈ¢ò `Êó©Ëµ∑Python`

````python
df.plot(figsize=(10,6),title='Êó©Ëµ∑Python')
plt.show()
````

## Ê∑ªÂä†ÁΩëÊ†º

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÁªôÊäòÁ∫øÂõæÊ∑ªÂä†ÁΩëÊ†º

````python
df.plot(figsize=(10,6),title='Êó©Ëµ∑Python',grid = True)
plt.show()
````

## Ê∑ªÂä†ËΩ¥Ê†áÁ≠æ

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÁªôXËΩ¥Ê∑ªÂä†Ê†áÁ≠æÔºàÊó∂Èó¥ÔºâÔºåYËΩ¥Ê∑ªÂä†Ê†áÁ≠æÔºàÊï∞ÈáèÔºâ

````python
df.plot(figsize=(10, 6), title='Êó©Ëµ∑Python', grid=True, xlabel='Êó∂Èó¥', ylabel='Êï∞Èáè')
plt.show()
````

## Ë∞ÉÊï¥ÂàªÂ∫¶Â§ßÂ∞è

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÂ∞ÜÂàªÂ∫¶Ë∞ÉÂ§ß‰∏ÄÁÇπ

````python
df.plot(figsize=(10, 6), title='Êó©Ëµ∑Python', grid=True, xlabel='Êó∂Èó¥', ylabel='Êï∞Èáè',fontsize = 13)
plt.show()
````

## Ë∞ÉÊï¥ÊñáÂ≠óÂ§ßÂ∞è

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÂ∞ÜÊ†áÈ¢ò„ÄÅÂùêÊ†áËΩ¥ÊñáÂ≠óÈÄÇÂΩìË∞ÉÂ§ß

````python
df.plot(figsize=(10, 6), grid=True, fontsize = 13)
plt.title("Êó©Ëµ∑Python",size = 17)
plt.xlabel('Êó∂Èó¥',size = 15)
plt.ylabel('Êï∞Èáè',size =15)
plt.show()
````

## Âõæ‰æã‰ΩçÁΩÆ

Âú®‰∏ä‰∏ÄÈ¢òÁöÑÂü∫Á°Ä‰∏äÔºåÂ∞ÜÂõæ‰æã‰ΩçÁΩÆË∞ÉÊï¥Âà∞Â∑¶‰∏ãËßí

````python
df.plot(figsize=(10, 6), grid=True, fontsize = 13)
plt.title("Êó©Ëµ∑Python",size = 17)
plt.xlabel('Êó∂Èó¥',size = 15)
plt.ylabel('Êï∞Èáè',size =15)
plt.legend(loc=3)
plt.show()
````

## ÂèåyËΩ¥

A„ÄÅB‰ΩøÁî®‰∏Ä‰∏™yËΩ¥ÔºåCD‰ΩøÁî®‰∏Ä‰∏™yËΩ¥

````python
ax = df.plot(secondary_y=['A', 'B'], figsize=(10, 6), fontsize = 13)
ax.set_ylabel('CD')  
ax.right_ax.set_ylabel('AB')  
ax.legend(loc=2)
plt.title("Êó©Ëµ∑Python",size = 17)
plt.xlabel('Êó∂Èó¥',size = 15)
plt.ylabel('Êï∞Èáè',size =15)
plt.legend(loc=1) 
plt.show()
````
